## Jawaban:
### 1.  Menampilkan data customer yg mempunyai opening amount diatas 7000.


```python
# Menampilkan Semua Data pada Tabel Customer
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM customer;"
result.execute(command)
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

#### Data customer yg mempunyai opening amount diatas 7000:

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM customer WHERE OPENING_AMT > '7000';"

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

### 2. Menampilkan informasi order yg dilakukan melalui agensi dengan code A003
#### Menampilkan semua data pada tabel orders

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM orders;"

# Eksekusi Query SQL
result.execute(command)

# Tampilkan Hasil dalam Bentuk DataFrame
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

#### Menampilkan informasi order yg dilakukan melalui agensi dengan code A003

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM orders WHERE AGENT_CODE = 'A003';"

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

### 3. Menampilkan data customer yg tinggal di New York, Bangalore, Mumbai, dan Torento

```python
# Buat Koneksi ke Database
connection = sqlite3.connect ('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM customer WHERE CUST_CITY IN ('New York', 'Bangalore', 'Mumbai', 'Torento');"

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

### 4. Menampilkan informasi customer yang mempunyai payment amount diatas 7000 dan urutkan berdasarkan opening amountnya dari yg terbesar

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM customer WHERE PAYMENT_AMT > '7000' ORDER BY OPENING_AMT DESC;"

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

### 5. Menampilkan data agensi yang tidak bekerja di london dan memiliki comission lebih dari 0.12

#### Menampilkan semua data pada tabel agents

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Query SQL
result = connection.cursor()
command = "SELECT * FROM agents;"

# Eksekusi Query SQL
result.execute(command)

# Tampilkan Hasil dalam Bentuk DataFrame
results = result.fetchall(); connection.close();
cols = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)
results_data
```

#### Menampilkan data agensi yang tidak bekerja di london dan memiliki comission lebih dari 0.12

```python
# Buat Koneksi ke Database
connection = sqlite3.connect('data_penjualan.db')

# Membuat Query SQL
result = connection.cursor()
command = "SELECT * FROM agents WHERE WORKING_AREA NOT IN ('London') AND COMMISSION > '0.12';"

# Eksekusi Query SQL
result.execute(command)

# Menampilkan Hasil Dalam Bentuk DataFrame
results      = result.fetchall();connection.close();
cols         = [column[0] for column in result.description]
results_data = pd.DataFrame.from_records(results, columns = cols)

results_data
```
